## üìö 1. Imports et Configuration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings

warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8-darkgrid')
%matplotlib inline

print("‚úÖ Imports r√©ussis")
```

## üìÇ 2. Chargement des Donn√©es

```python
df = pd.read_csv('../../donnees_cliniques_nettoyees.csv')
print(f"üìä {df.shape[0]} cliniques, {df.shape[1]} colonnes")
df.head(3)
```

## üìä 3. Analyse des Canaux Digitaux

```python
# Analyse canaux
canaux = {
    'uses_website': 'Site Web',
    'uses_social_media': 'R√©seaux Sociaux',
    'uses_newsletters': 'Newsletter'
}

canal_data = []
for col, label in canaux.items():
    if col in df.columns:
        count = df[col].apply(lambda x: 1 if str(x).lower() in ['1', '1.0', 'oui'] else 0).sum()
        pct = (count / len(df)) * 100
        canal_data.append({'Canal': label, 'Adoption': int(count), 'Pourcentage': pct})

df_canaux = pd.DataFrame(canal_data).sort_values('Pourcentage', ascending=False)
print("\nüìä ADOPTION DES CANAUX DIGITAUX")
print(df_canaux.to_string(index=False))

# Visualisation
fig = px.bar(df_canaux, x='Pourcentage', y='Canal', orientation='h',
             title='üì± Adoption des Canaux Digitaux',
             color='Pourcentage', color_continuous_scale='Blues',
             text='Pourcentage')
fig.update_traces(texttemplate='%{text:.1f}%', textposition='outside')
fig.write_html('reports/adoption_canaux.html')
fig.show()
```

## üìù 4. Analyse des Types de Contenu

```python
# Types de contenu (exemple avec r√©seaux sociaux)
contenu_columns = [
    'content_social_health_tips',
    'content_social_prof_articles',
    'content_social_patient_info',
    'content_social_services_reviews'
]

print("\nüìù TYPES DE CONTENU UTILIS√âS")
for col in contenu_columns:
    if col in df.columns:
        count = df[col].apply(lambda x: 1 if str(x).lower() in ['1', '1.0', 'oui'] else 0).sum()
        pct = (count / len(df)) * 100
        print(f"  {col}: {pct:.1f}%")
```

## üéØ 5. Objectifs de Communication

```python
# Objectifs
objectifs = {
    'goal_brand_awareness': 'Notori√©t√© de marque',
    'goal_patient_acquisition': 'Acquisition patients',
    'goal_patient_retention': 'Fid√©lisation'
}

obj_data = []
for col, label in objectifs.items():
    if col in df.columns:
        count = df[col].apply(lambda x: 1 if str(x).lower() in ['1', '1.0', 'oui'] else 0).sum()
        pct = (count / len(df)) * 100
        obj_data.append({'Objectif': label, 'Cliniques': int(count), 'Pourcentage': pct})

df_obj = pd.DataFrame(obj_data).sort_values('Pourcentage', ascending=False)
print("\nüéØ OBJECTIFS DE COMMUNICATION")
print(df_obj.to_string(index=False))
df_obj.to_csv('data/objectifs_communication.csv', index=False)
print("\n‚úÖ Donn√©es export√©es")
```
