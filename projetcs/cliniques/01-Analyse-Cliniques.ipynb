### **Topo des Données et de l'Analyse de la Maturité Digitale**

#### **1. Présentation des Données**

*   **Origine :** Le jeu de données provient d'un fichier Excel nommé `EVALUATION_DES_BESOINS_DES_CLINIQUES_PRIVEES_-_all_versions_-_labels_-_2025-12-17-15-27-15.xlsx`. Il contient les réponses à un questionnaire envoyé à des **cliniques privées**, probablement dans un but d'évaluation de leurs besoins en matière de numérisation et de gestion.
*   **Contenu :** Les données comprennent des informations descriptives sur les cliniques (nom, localisation, nombre de lits, de professionnels, spécialités), leurs pratiques actuelles en matière de **gestion administrative** (systèmes d'information, tâches chronophages), de **communication digitale** (présence web, réseaux sociaux, newsletters, stratégie), d'**automatisation** souhaitée, et d'**intelligence artificielle** (intérêt, freins, bénéfices attendus).
*   **Structure :** Initialement, les colonnes ont des noms longs et verbeux, provenant d'un formulaire de type KoBo Toolbox. Elles ont été **renomées** pour plus de clarté et de concision (ex : `DISPOSEZ-VOUS D'UN SYSTEME DE GESTION INFORMATISE?` devient `has_informatic_management_system`).
*   **Format :** Le fichier a été **nettoyé** (valeurs manquantes standardisées, espaces supprimés, types de données corrigés) et sauvegardé au format **CSV** (`donnees_cliniques_nettoyees.csv`) pour faciliter l'analyse.

#### **2. Objectif de l'Analyse : Cartographier la Maturité Digitale**

*   **But :** Évaluer **à quel point les cliniques privées répondantes sont matures sur le plan numérique**. Cela signifie mesurer dans quelle mesure elles utilisent des outils, canaux et processus numériques pour gérer leur activité, communiquer et potentiellement innover.
*   **Approche :** Analyser des indicateurs concrets tirés des réponses du questionnaire, tels que la possession d'un système de gestion informatisé, la présence en ligne, l'utilisation d'outils numériques pour les tâches administratives, etc.

#### **3. KPIs Sélectionnés et Justification**

Nous avons défini un ensemble de KPIs pour quantifier la maturité digitale :

*   **KPI 1 : Pourcentage de cliniques disposant de...**
    *   `has_informatic_management_system` (Système de gestion informatisé)
    *   `uses_website` (Site web)
    *   `uses_social_media` (Réseaux sociaux)
    *   `uses_digital_tools_for_appointments` (Outils numériques pour les RDV)
    *   **Pourquoi ?** Ces éléments sont des piliers fondamentaux de la présence et de la gestion numérique. Leur absence indique souvent un niveau de maturité faible.
    *   **Comment interpréter ?** Un pourcentage élevé (ex : >70%) signifie que la majorité des cliniques interrogées ont adopté cet outil de base. Un pourcentage faible (ex : <30%) indique une opportunité de croissance significative dans ce domaine.

*   **KPI 2 : Nombre moyen de canaux digitaux utilisés par clinique (`num_digital_channels`)**
    *   **Calcul :** Somme des 4 indicateurs binaires ci-dessus pour chaque clinique.
    *   **Pourquoi ?** Permet d'avoir une vue d'ensemble agrégée de l'adoption numérique. Une clinique avec un score de 3/4 est plus "connectée" qu'une clinique avec 1/4.
    *   **Comment interpréter ?** Un score moyen proche de 4 signifie que les cliniques sont globalement très actives numériquement. Un score moyen proche de 0 ou 1 indique une faible adoption globale. Un score moyen intermédiaire (ex : 1.5 à 2.5) suggère une adoption partielle, peut-être limitée à un ou deux outils essentiels.

*   **KPI 3 : Comparaison par taille de clinique (`size_category`)**
    *   **Analyse :** Comparer les pourcentages des indicateurs ci-dessus et le `num_digital_channels` moyen entre les "Petites cliniques" (≤ X lits) et les "Grandes cliniques" (> X lits).
    *   **Pourquoi ?** Les grandes structures ont souvent plus de ressources (humaines, financières) pour investir dans la numérisation. Cette comparaison permet d'identifier d'éventuels **écarts de maturité** liés à la taille, ce qui peut orienter des stratégies de soutien différenciées.
    *   **Comment interpréter ?** Si les grandes cliniques montrent des taux d'adoption nettement supérieurs, cela confirme l'hypothèse liée à la taille. Si les écarts sont faibles, cela suggère que la taille n'est pas un facteur déterminant majeur dans ce contexte.

*   **KPI 4 : Score de maturité digitale agrégé (`maturity_score`)**
    *   **Calcul :** Basé sur la somme des 4 indicateurs binaires (max 4) + potentiellement un point bonus pour l'utilisation de plus de 2 canaux (max 1), donnant un score de 0 à 5.
    *   **Pourquoi ?** Offrir un **indicateur synthétique** simple pour chaque clinique. Cela facilite la catégorisation (ex : "Très mature (4-5)", "Moyennement mature (2-3)", "Faiblement mature (0-1)") et la comparaison globale.
    *   **Comment interpréter ?** Chaque clinique se retrouve dans une classe de maturité. Une distribution du score permet de visualiser la répartition de la maturité au sein de l'échantillon. Beaucoup de cliniques avec un score bas signifie un fort potentiel de numérisation.

#### **4. Utilité de l'Analyse**

*   **Diagnostic :** Fournir un état des lieux objectif de la situation actuelle des cliniques en matière de numérisation.
*   **Identification des opportunités :** Mettre en lumière les domaines négligés (ex : absence de site web) ou les freins communs (ex : manque de compétences).
*   **Segmentation :** Permettre de segmenter les cliniques selon leur niveau de maturité, pour proposer des offres ou des accompagnements personnalisés (ex : formation de base pour les moins matures, solutions avancées pour les plus matures).
*   **Argumentaire de vente/développement :** Servir de base factuelle pour justifier la pertinence de solutions numériques ou d'accompagnement à destination des cliniques ou des parties prenantes du secteur de la santé.
*   **Suivi d'évolution :** En réitérant l'analyse à intervalles réguliers, mesurer l'évolution de la maturité digitale du secteur privé.

```python
import pandas as pd
import numpy as np

print("DataFrame avant nettoyage:")
print(df_renamed.info())
print("\nPremières lignes:")
print(df_renamed.head())

# --- 1. Types de données ---

# Colonnes datetime
datetime_cols = ['survey_start_datetime', 'survey_end_datetime', 'submission_time']
for col in datetime_cols:
    if col in df_renamed.columns:
        df_renamed[col] = pd.to_datetime(df_renamed[col], errors='coerce') # errors='coerce' met les valeurs impossibles en NaT

# Colonnes numériques potentielles
numeric_cols = ['number_of_beds', 'number_of_healthcare_staff']
for col in numeric_cols:
    if col in df_renamed.columns:
        df_renamed[col] = pd.to_numeric(df_renamed[col], errors='coerce') # errors='coerce' met les valeurs impossibles en NaN

# --- 2. & 3. Nettoyage du texte et gestion des valeurs manquantes ---
# Valeurs considérées comme manquantes
na_values_list = ['', ' ', 'NA', 'N/A', '-', 'Aucun', 'Aucune', 'Rien', 'Rien à Signaler', 'RAS', 'Zéro', 
                  '00 fois', '0', 'Aucune équipe', 'Aucune stratégie', 'Aucune difficulté', 'Aucune implication',
                  'Aucun retour', 'Aucune veille', 'Aucune adaptation', 'Aucune barrière', 'Aucun frein', 'Aucun support', 
                  'Aucun indicateur', 'Aucune attente', 'Aucun personnel', 'Aucun problème', 'Aucun impact', 'Aucune perception',
                  'Aucune suggestion', 'Aucune idée', 'Pas encore', 'Pas impliquées', 'Pas de communication', 'Pas de stratégie',
                  'Pas de budget', 'Pas de personnel', 'Pas de compétence', 'Pas de solution', 'Pas de plan', 'Pas de retour',
                  'Pas de veille', 'Pas d\'indicateur', 'Pas d\'attente', 'Pas de suggestion', 'Rien à signaler', 'Non applicable', 
                  'Non défini', 'ND', 'N.D.', 'NULL', 'null']

# Remplacer les valeurs NA par pd.NA pour toutes les colonnes
df_renamed.replace(na_values_list, pd.NA, inplace=True)

# Supprimer les espaces blancs en début/fin pour les colonnes de type objet (str)
object_cols = df_renamed.select_dtypes(include=['object']).columns
df_renamed[object_cols] = df_renamed[object_cols].apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# --- 4. Normalisation des colonnes binaires (1/0) ---
# Sélectionner les colonnes binaires (celles qui ont des valeurs 0 ou 1 ou NA)
# On suppose que les colonnes binaires contiennent des entiers ou des chaînes '0'/'1'
binary_like_cols = [
    # Tâches admin
    'time_consuming_appointments', 'time_consuming_patient_records', 'time_consuming_billing_payment', 'time_consuming_inventory_orders', 'time_consuming_other_flag',
    # Défis dossiers
    'challenge_data_centralization', 'challenge_accessing_info', 'challenge_data_entry_errors', 'challenge_other_patient_records_flag',
    # Processus à automatiser
    'want_automation_appointments', 'want_automation_patient_records', 'want_automation_billing', 'want_automation_data_analysis', 'want_automation_inventory', 'want_automation_other_flag',
    # Canaux digitaux
    'uses_website', 'uses_social_media', 'uses_newsletters', 'uses_other_channel_flag',
    # Contenu Site Web
    'content_website_health_tips', 'content_website_prof_articles', 'content_website_deep_dive', 'content_website_patient_info', 'content_website_services_reviews', 'content_website_other_flag',
    # Contenu Réseaux Sociaux
    'content_social_health_tips', 'content_social_prof_articles', 'content_social_deep_dive', 'content_social_patient_info', 'content_social_services_reviews', 'content_social_other_flag',
    # Contenu Newsletters
    'content_newsletter_health_tips', 'content_newsletter_prof_articles', 'content_newsletter_deep_dive', 'content_newsletter_patient_info', 'content_newsletter_services_reviews', 'content_newsletter_other_flag',
    # Objectifs communication
    'goal_brand_awareness', 'goal_patient_acquisition', 'goal_patient_retention', 'goal_other_comm_flag',
    # Attentes IA
    'expect_IA_admin_efficiency', 'expect_IA_accuracy_improvement', 'expect_IA_resource_optimization', 'expect_IA_better_patient_experience', 'expect_IA_other_benefit_flag',
    # Freins IA
    'barrier_IA_cost', 'barrier_IA_lack_of_skills', 'barrier_IA_security_privacy', 'barrier_IA_resistance_change', 'barrier_IA_other_flag',
    # Support IA
    'support_IA_training', 'support_IA_tech_assistance', 'support_IA_digital_consulting', 'support_IA_other_flag'
]

for col in binary_like_cols:
    if col in df_renamed.columns:
        # Convertir en entier après avoir remplacé les '1'/'0' string par des ints
        df_renamed[col] = pd.to_numeric(df_renamed[col], errors='coerce').astype('Int8') # Int8 permet les NA

# --- 5. Création de nouvelles features (exemple) ---
if 'survey_end_datetime' in df_renamed.columns and 'survey_start_datetime' in df_renamed.columns:
    df_renamed['survey_duration_seconds'] = (df_renamed['survey_end_datetime'] - df_renamed['survey_start_datetime']).dt.total_seconds()
if 'submission_time' in df_renamed.columns and df_renamed['submission_time'].notna().any():
    df_renamed['submission_year'] = df_renamed['submission_time'].dt.year
    df_renamed['submission_month'] = df_renamed['submission_time'].dt.month
    df_renamed['submission_day_of_week'] = df_renamed['submission_time'].dt.dayofweek


print("\n\nDataFrame après nettoyage et normalisation:")
print(df_renamed.info())
print("\nPremières lignes:")
print(df_renamed.head())

# --- 6. Validation rapide ---
print("\n\nValeurs manquantes par colonne après nettoyage:")
print(df_renamed.isnull().sum())

# Exemple de vérification pour une colonne binaire
binary_col_example = 'uses_website'
if binary_col_example in df_renamed.columns:
    print(f"\nValeurs uniques pour {binary_col_example}:")
    print(df_renamed[binary_col_example].value_counts(dropna=False))

# Sauvegarde du DataFrame nettoyé (optionnel)
output_cleaned_file_path = 'donnees_cliniques_nettoyees.csv'
df_renamed.to_csv(output_cleaned_file_path, index=False, encoding='utf-8')
print(f"\nLe DataFrame nettoyé a été sauvegardé sous '{output_cleaned_file_path}'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assurez-vous que df_renamed est le DataFrame nettoyé
# df_renamed = ...

# --- 1. Préparation des données pour l'analyse de maturité ---

# Colonnes binaires pertinentes pour la maturité digitale
# Ces colonnes doivent être de type numérique (Int8) après le nettoyage
digital_maturity_cols = [
    'has_informatic_management_system', # 1 = Oui, 0 = Non (ou NA)
    'uses_website',                     # 1 = Oui, 0 = Non (ou NA)
    'uses_social_media',                # 1 = Oui, 0 = Non (ou NA)
    'uses_digital_tools_for_appointments' # 1 = Oui, 0 = Non (ou NA)
]

# Vérifiez que les colonnes existent et sont numériques
for col in digital_maturity_cols:
    if col not in df_renamed.columns:
        print(f"Attention : La colonne '{col}' est manquante.")
    elif not pd.api.types.is_numeric_dtype(df_renamed[col]):
         print(f"Attention : La colonne '{col}' n'est pas numérique.")

# Remplacer 'Oui'/'Non' par 1/0 si nécessaire (devrait être fait dans le nettoyage)
# Exemple pour 'has_informatic_management_system' si elle est encore en str
if df_renamed['has_informatic_management_system'].dtype == 'object':
    df_renamed['has_informatic_management_system'] = df_renamed['has_informatic_management_system'].map({'Oui': 1, 'Non': 0, 'Partiellement (mixte entre papier et numérique)': 1}).astype('Int8')

if df_renamed['uses_digital_tools_for_appointments'].dtype == 'object':
    df_renamed['uses_digital_tools_for_appointments'] = df_renamed['uses_digital_tools_for_appointments'].map({'Oui': 1, 'Non': 0}).astype('Int8')
```

```python

# --- 2. Calcul des indicateurs simples ---

# Pourcentage de cliniques avec chaque fonctionnalité
percentages = {}
for col in digital_maturity_cols:
    # Calcul du pourcentage en ne tenant compte que des valeurs non nulles
    total_non_null = df_renamed[col].notna().sum()
    if total_non_null > 0:
        percentage = (df_renamed[col].sum() / total_non_null) * 100
    else:
        percentage = 0 # Si toutes les valeurs sont nulles
    percentages[col] = percentage

print("\n--- Pourcentages de cliniques avec chaque fonctionnalité ---")
for k, v in percentages.items():
    print(f"{k}: {v:.2f}%")

# Calcul du nombre de canaux digitaux utilisés par clinique
# Somme des colonnes binaires pour chaque ligne (axis=1)
df_renamed['num_digital_channels'] = df_renamed[digital_maturity_cols].sum(axis=1, skipna=True)

mean_channels = df_renamed['num_digital_channels'].mean()
print(f"\nNombre moyen de canaux digitaux utilisés par clinique : {mean_channels:.2f}")
```

```python

# --- 3. Création des graphiques ---

# Configurer la grille de sous-graphiques
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Analyse de la Maturité Digitale des Cliniques', fontsize=16)

# --- 3.1. Barplot : Pourcentage de cliniques avec chaque fonctionnalité ---
ax1 = axes[0, 0]
# Préparer les données pour le barplot
plot_data = pd.Series(percentages)
plot_data.plot(kind='bar', ax=ax1, color='skyblue', edgecolor='black')
ax1.set_title('Pourcentage de Cliniques avec Fonctionnalités Numériques')
ax1.set_ylabel('Pourcentage (%)')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# --- 3.2. Histogramme : Distribution du nombre de canaux digitaux ---
ax2 = axes[0, 1]
df_renamed['num_digital_channels'].hist(bins=range(0, len(digital_maturity_cols) + 2), ax=ax2, color='lightgreen', edgecolor='black')
ax2.set_title('Distribution du Nombre de Canaux Digitaux Utilisés')
ax2.set_xlabel('Nombre de Canaux Digitaux')
ax2.set_ylabel('Nombre de Cliniques')
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# --- 3.3. Comparaison Petites vs Grandes Cliniques (ex : seuil à 30 lits) ---
# Définir un seuil pour "petite" vs "grande" clinique (vous pouvez ajuster)
bed_threshold = 30
df_renamed['size_category'] = df_renamed['number_of_beds'].apply(lambda x: 'Petite (<=30 lits)' if pd.notna(x) and x <= bed_threshold else 'Grande (>30 lits)')

# Calcul des pourcentages par catégorie de taille
comparison_data = df_renamed.groupby('size_category')[digital_maturity_cols].apply(
    lambda x: (x.sum() / x.count()) * 100 if x.count().sum() > 0 else pd.Series([0]*len(digital_maturity_cols), index=digital_maturity_cols)
).T

ax3 = axes[1, 0]
comparison_data.plot(kind='bar', ax=ax3, width=0.8)
ax3.set_title('Comparaison de la Maturité Numérique\n(Petites vs Grandes Cliniques)')
ax3.set_ylabel('Pourcentage (%)')
ax3.legend(title='Catégorie')
ax3.tick_params(axis='x', rotation=45)
ax3.grid(axis='y', linestyle='--', alpha=0.7)

# --- 3.4. Comparaison Petites vs Grandes Cliniques - Nombre moyen de canaux ---
mean_channels_by_size = df_renamed.groupby('size_category')['num_digital_channels'].mean()

ax4 = axes[1, 1]
mean_channels_by_size.plot(kind='bar', ax=ax4, color='orange', edgecolor='black')
ax4.set_title('Nombre Moyen de Canaux Digitaux\npar Taille de Clinique')
ax4.set_ylabel('Nombre Moyen de Canaux')
ax4.tick_params(axis='x', rotation=45)
ax4.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout(rect=[0, 0, 1, 0.96]) # Ajuste la disposition pour le titre global
plt.show()

# --- 4. Calcul d'un score simple de maturité digitale (0-5) ---
# On attribue un point par fonctionnalité présente (max 4) + un point si > 2 canaux utilisés (max 1)
# Vous pouvez ajuster la logique du score selon vos critères
max_features_score = 4 # 1 point par fonctionnalité
df_renamed['maturity_score_features'] = df_renamed[digital_maturity_cols].sum(axis=1, min_count=1) # min_count=1 pour gérer les lignes avec que des NA
df_renamed['maturity_score_channels'] = df_renamed['num_digital_channels'].apply(lambda x: 1 if pd.notna(x) and x > 2 else 0)
df_renamed['maturity_score'] = df_renamed['maturity_score_features'] + df_renamed['maturity_score_channels']

# Remplacer les scores NaN (lignes où aucune donnée n'était disponible) par 0 ou une autre valeur par défaut
df_renamed['maturity_score'] = df_renamed['maturity_score'].fillna(0)

# Afficher des statistiques sur le score
print("\n--- Statistiques du Score de Maturité Digitale (0-5) ---")
print(df_renamed['maturity_score'].describe())
print("\nRépartition des scores :")
print(df_renamed['maturity_score'].value_counts().sort_index())

# Afficher un histogramme du score
plt.figure(figsize=(10, 6))
df_renamed['maturity_score'].hist(bins=range(0, 7), edgecolor='black', color='lightcoral')
plt.title('Distribution du Score de Maturité Digitale (0-5)')
plt.xlabel('Score de Maturité')
plt.ylabel('Nombre de Cliniques')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
