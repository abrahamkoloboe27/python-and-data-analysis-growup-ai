## 🚨 La Problématique : Pourquoi a-t-on besoin d'environnements virtuels ?

Imagine la situation suivante :

```
Projet A (2022) : pandas==1.3.0, numpy==1.21.0
Projet B (2024) : pandas==2.1.0, numpy==1.26.0
```

Si tu travailles sur les deux projets sur le même ordinateur **sans isolation**, tu vas avoir des conflits !

### 😱 Problèmes sans environnements virtuels :

| Problème | Description |
|----------|-------------|
| **Conflits de versions** | Le Projet A casse quand tu mets à jour pandas pour le Projet B |
| **Pollution de l'environnement global** | Des centaines de packages installés partout |
| **Non-reproductibilité** | `"Ça marche chez moi mais pas chez toi"` |
| **Problèmes de permissions** | Modifier Python système peut casser l'OS |
| **Difficultés de collaboration** | Chaque dev a des versions différentes |

```
          🖥️ Ton ordinateur (SANS environnements virtuels)
          ┌─────────────────────────────────────────────┐
          │  Python 3.10 GLOBAL                         │
          │  ├── pandas 2.1.0   ← Projet B veut ça     │
          │  ├── numpy 1.26.0   ← Projet B veut ça     │
          │  └── ...                                    │
          │                                             │
          │  ❌ Projet A : pandas 1.3.0 ← CONFLIT !    │
          └─────────────────────────────────────────────┘
```

## 1️⃣ Qu'est-ce qu'un Environnement Virtuel ?

Un **environnement virtuel** est un **dossier isolé** qui contient :
- Sa propre copie de l'interpréteur Python
- Ses propres packages installés
- Sa propre configuration

Il est **complètement indépendant** du Python installé sur le système et des autres environnements.

```
          🖥️ Ton ordinateur (AVEC environnements virtuels)
          ┌─────────────────────────────────────────────────┐
          │  Python 3.10 GLOBAL                             │
          │                                                 │
          │  📁 venv_projet_a/    📁 venv_projet_b/        │
          │  ┌──────────────┐    ┌──────────────────────┐  │
          │  │ Python 3.10  │    │ Python 3.11          │  │
          │  │ pandas 1.3.0 │    │ pandas 2.1.0         │  │
          │  │ numpy 1.21.0 │    │ numpy 1.26.0         │  │
          │  └──────────────┘    └──────────────────────┘  │
          │      ✅ OK               ✅ OK                  │
          └─────────────────────────────────────────────────┘
```

### 🔑 Principes fondamentaux

1. **Isolation** : Chaque projet a ses propres packages, sans interférence
2. **Légèreté** : Un environnement virtuel n'est qu'un dossier (pas une VM !)
3. **Portabilité** : On peut décrire un environnement via `requirements.txt`
4. **Reproductibilité** : N'importe qui peut recréer exactement le même environnement

## 2️⃣ pip — Le Gestionnaire de Packages Python

**pip** (*Pip Installs Packages*) est l'outil standard pour installer des packages Python depuis [PyPI](https://pypi.org) (Python Package Index).

### 🔹 Vérifier la version de pip

```bash
pip --version
# ou
pip3 --version
```

### 🔹 Commandes de base

```bash
# Installer un package
pip install numpy

# Installer une version spécifique
pip install numpy==1.26.0

# Installer une version minimale
pip install numpy>=1.20.0

# Mettre à jour un package
pip install --upgrade numpy

# Désinstaller un package
pip uninstall numpy

# Lister les packages installés
pip list

# Voir les informations d'un package
pip show pandas

# Chercher un package (sur PyPI via navigateur ou API)
# ⚠️ 'pip search' est désactivé depuis 2021 — utilise https://pypi.org à la place
```

### 🔹 Le fichier `requirements.txt`

Ce fichier liste **toutes les dépendances** du projet, avec les versions.

```bash
# Générer requirements.txt depuis l'environnement actuel
pip freeze > requirements.txt

# Installer toutes les dépendances d'un projet
pip install -r requirements.txt
```

**Exemple de `requirements.txt` :**

```
numpy==1.26.0
pandas==2.1.0
scikit-learn==1.3.2
matplotlib==3.8.0
jupyter==1.0.0
```

### ⚠️ Limites de pip seul

- pip n'isole **pas** les environnements entre projets
- `pip freeze` capture **tous** les packages (même ceux non nécessaires)
- Pas de gestion automatique des **conflits de dépendances**
- Ne gère pas les **versions de Python**

## 3️⃣ venv — Créer des Environnements Virtuels

**venv** est le module Python standard (inclus depuis Python 3.3) pour créer des environnements virtuels. Pas besoin d'installer quoi que ce soit !

### 🔹 Créer un environnement virtuel

```bash
# Syntaxe de base
python -m venv nom_de_l_environnement

# Convention : nommer l'environnement 'venv' ou '.venv'
python -m venv venv

# Avec une version spécifique de Python
python3.11 -m venv venv
```

**Structure créée :**

```
mon_projet/
├── venv/
│   ├── bin/          (Linux/Mac) ou Scripts/ (Windows)
│   │   ├── python    ← copie de l'interpréteur Python
│   │   ├── pip       ← pip dédié à cet env
│   │   └── activate  ← script d'activation
│   ├── lib/
│   │   └── python3.x/
│   │       └── site-packages/  ← packages installés ici
│   └── include/
├── main.py
└── requirements.txt
```

### 🔹 Activer l'environnement virtuel

```bash
# 🐧 Linux / 🍎 macOS
source venv/bin/activate

# 🪟 Windows (Command Prompt)
venv\Scripts\activate.bat

# 🪟 Windows (PowerShell)
venv\Scripts\Activate.ps1
```

Après activation, ton terminal affiche le nom de l'environnement :

```bash
(venv) user@machine:~/mon_projet$ 
#  ↑ Indique que l'environnement est actif
```

### 🔹 Installer des packages dans l'environnement

```bash
# Après activation, pip installe dans le venv, pas globalement !
(venv) $ pip install pandas numpy matplotlib

# Vérifier ce qui est installé dans CET environnement
(venv) $ pip list

# Figer les versions
(venv) $ pip freeze > requirements.txt
```

### 🔹 Désactiver l'environnement

```bash
(venv) $ deactivate
# Le préfixe (venv) disparaît
user@machine:~/mon_projet$
```

### 🔹 Supprimer un environnement

```bash
# Un environnement virtuel = juste un dossier !
# Linux/Mac
rm -rf venv/

# Windows
rmdir /s /q venv
```

### 📌 Bonne pratique : `.gitignore`

Ne jamais committer le dossier `venv/` dans Git ! Ajoute-le à `.gitignore` :

```
# .gitignore
venv/
.venv/
__pycache__/
*.pyc
```

💡 **À la place**, committe `requirements.txt` pour que chacun puisse recréer l'environnement.

````python
# Démonstration : afficher la version de Python et les packages installés
import sys
import subprocess

print(f"🐍 Version Python : {sys.version}")
print(f"📍 Chemin de l'interpréteur : {sys.executable}")
print()

# Lister les packages installés
result = subprocess.run(
    [sys.executable, "-m", "pip", "list", "--format=columns"],
    capture_output=True,
    text=True
)
print("📦 Packages installés dans cet environnement :")
print(result.stdout)
````

````python
# Vérifier si on est dans un environnement virtuel
import sys
import os

def est_dans_venv():
    """Vérifie si Python tourne dans un environnement virtuel."""
    # sys.prefix != sys.base_prefix quand on est dans un venv
    return sys.prefix != sys.base_prefix

if est_dans_venv():
    print("✅ Tu es dans un environnement virtuel !")
    print(f"   📁 Dossier de l'env : {sys.prefix}")
else:
    print("⚠️  Tu utilises Python global (pas de venv actif)")
    print(f"   📍 Python global : {sys.base_prefix}")

# Vérifier la variable d'environnement VIRTUAL_ENV
venv_path = os.environ.get('VIRTUAL_ENV')
if venv_path:
    print(f"\n🔍 Variable VIRTUAL_ENV : {venv_path}")
````

## 5️⃣ L'Importance en Data Science

Les environnements virtuels sont **encore plus critiques** en data science. Voici pourquoi :

---

### 🔬 1. Reproductibilité des Expériences

En data science, la **reproductibilité** est fondamentale. Un modèle ML entraîné avec `scikit-learn==1.2.0` peut donner des résultats **différents** avec `scikit-learn==1.3.0` !

```
❌ Sans venv :
   Collègue A : scikit-learn 1.2.0 → Accuracy: 87.3%
   Collègue B : scikit-learn 1.3.0 → Accuracy: 85.1%  ← Même code, résultat différent !

✅ Avec venv + requirements.txt figé :
   Collègue A : scikit-learn==1.2.0 → Accuracy: 87.3%
   Collègue B : scikit-learn==1.2.0 → Accuracy: 87.3%  ← Résultat identique ✓
```

---

### 🐍 2. Gestion des Versions de Python

Certaines bibliothèques data ne supportent que certaines versions de Python :

| Bibliothèque | Python supporté |
|-------------|----------------|
| TensorFlow 2.13 | Python 3.8 – 3.11 |
| PyTorch 2.1 | Python 3.8 – 3.11 |
| pandas 2.x | Python 3.9+ |
| Ancien projet (2020) | Python 3.7 |

```bash
# Créer un env avec une version Python spécifique
python3.9 -m venv venv_projet_ancien
python3.11 -m venv venv_projet_recent
```

---

### 📌 3. Figer les Versions des Packages

**Figer** = enregistrer les versions exactes utilisées, pour les reproduire plus tard.

```bash
# Méthode 1 : pip freeze (capture TOUT ce qui est installé)
pip freeze > requirements.txt

# Exemple de sortie :
# numpy==1.26.0
# pandas==2.1.0
# python-dateutil==2.8.2
# pytz==2023.3
# six==1.16.0
# tzdata==2023.3
```

💡 Conseil : utiliser des **versions exactes** (`==`) en production, des **versions minimales** (`>=`) pour les librairies open-source.

---

### 🌳 4. Gestion de l'Arbre de Dépendances

Chaque package a ses propres **dépendances transitives** qui peuvent créer des conflits.

```
Exemple d'arbre de dépendances :

pandas
├── numpy>=1.23.0
├── python-dateutil>=2.8.2
│   └── six>=1.5
├── pytz>=2020.1
└── tzdata>=2022.7

matplotlib
├── numpy>=1.21
├── pillow>=8.0
├── packaging>=20.0
└── ...
```

Problème possible :
- `package_A` requiert `numpy>=1.21, <1.25`
- `package_B` requiert `numpy>=1.24`
- → Conflit ! Impossible de satisfaire les deux !

```bash
# Visualiser l'arbre de dépendances
pip install pipdeptree
pipdeptree

# Vérifier les conflits
pip check
```

En data science, le code doit souvent tourner sur :
- Des serveurs de production

````python
# Démonstration : visualiser les dépendances d'un package
import importlib.metadata

def afficher_dependances(package_name):
    """Affiche les dépendances directes d'un package installé."""
    try:
        dist = importlib.metadata.distribution(package_name)
        requires = dist.metadata.get_all('Requires-Dist') or []
        version = dist.metadata['Version']
        
        print(f"📦 {package_name} v{version}")
        if requires:
            print(f"   Dépendances ({len(requires)}) :")
            for dep in requires[:10]:  # Limiter à 10 pour la lisibilité
                print(f"   ├── {dep}")
            if len(requires) > 10:
                print(f"   └── ... et {len(requires) - 10} autres")
        else:
            print("   Aucune dépendance externe")
    except importlib.metadata.PackageNotFoundError:
        print(f"❌ Package '{package_name}' non installé")

# Tester avec quelques packages data
for pkg in ["pandas", "numpy", "matplotlib"]:
    afficher_dependances(pkg)
    print()
````

## 6️⃣ uv — Le Gestionnaire Moderne qui Simplifie Tout

**`uv`** est un outil révolutionnaire créé par [Astral](https://astral.sh) (les créateurs de `ruff`).
Il est écrit en **Rust** 🦀, ce qui le rend **extrêmement rapide** — jusqu'à **100x plus rapide** que pip !

### 🤔 Pourquoi uv change la donne ?

Avant `uv`, gérer un projet Python nécessitait **plusieurs outils** :

```
AVANT uv :
  pyenv    → gérer les versions de Python
+ venv     → créer l'environnement virtuel
+ pip      → installer les packages
+ pip-tools → figer les dépendances
= 4 outils, syntaxes différentes, fichiers de config multiples

AVEC uv :
  uv       → fait tout ça et plus, en une seule commande !
```

### ⚡ Comparatif de performances

| Opération | pip | uv |
|-----------|-----|----|
| Installer numpy | ~8s | ~0.1s |
| Installer pandas + dépendances | ~45s | ~2s |
| Résoudre les conflits | Lent | Instantané |

> 💡 `uv` utilise un résolveur de dépendances ultra-optimisé écrit en Rust

### 🔧 Installation de uv

```bash
# Méthode recommandée (Linux/Mac)
# 💡 Conseil sécurité : télécharge le script, lis-le, puis exécute-le
curl -LsSf https://astral.sh/uv/install.sh -o install.sh
less install.sh    # ou: cat install.sh
sh install.sh

# Option équivalente mais déconseillée (exécute du code distant non audité) :
# curl -LsSf https://astral.sh/uv/install.sh | sh

# Windows
powershell -ExecutionPolicy ByPass -c "irm https://astral.sh/uv/install.ps1 | iex"

# Via pip (si tu as déjà Python)
pip install uv

# Vérifier l'installation
uv --version
```

### 🚀 Démarrer un projet avec uv

```bash
# Initialiser un nouveau projet
uv init mon_projet_data
cd mon_projet_data
```

**Structure créée automatiquement :**

```
mon_projet_data/
├── .python-version    ← version Python fixée automatiquement
├── pyproject.toml     ← configuration du projet (remplace setup.py)
├── uv.lock            ← fichier de verrouillage des dépendances
├── README.md
└── main.py
```

### 🐍 Gérer les versions de Python avec uv

```bash
# Lister les versions Python disponibles
uv python list

# Installer une version de Python
uv python install 3.11
uv python install 3.12

# Définir la version Python pour le projet
uv python pin 3.11
# → Crée/met à jour .python-version
```

### 📦 Gérer les packages avec uv

```bash
# Ajouter un package (met à jour pyproject.toml + uv.lock)
uv add pandas
uv add "numpy>=1.26"
uv add pandas numpy scikit-learn matplotlib

# Ajouter une dépendance de développement
uv add --dev pytest black ruff

# Supprimer un package
uv remove pandas

# Mettre à jour un package
uv add --upgrade pandas

# Installer toutes les dépendances (depuis uv.lock)
uv sync

# Lister les packages installés
uv pip list
```

### ▶️ Exécuter du code avec uv

```bash
# Lancer un script (active l'env automatiquement)
uv run python main.py

# Lancer un module
uv run python -m pytest

# Lancer un notebook Jupyter
uv run jupyter notebook

# Plus besoin de 'source venv/bin/activate' !
```

### 📄 Le fichier `pyproject.toml` avec uv

`uv` utilise `pyproject.toml` (standard moderne Python) pour configurer le projet :

```toml
[project]
name = "mon-projet-data"
version = "0.1.0"
description = "Analyse des données de ventes"
requires-python = ">=3.11"
dependencies = [
    "pandas>=2.1.0",
    "numpy>=1.26.0",
    "scikit-learn>=1.3.0",
    "matplotlib>=3.8.0",
    "plotly>=5.18.0",
    "jupyter>=1.0.0",
]

[tool.uv]
dev-dependencies = [
    "pytest>=7.4.0",
    "black>=23.0.0",
    "ruff>=0.1.0",
]
```

### 🔒 Le fichier `uv.lock` — Verrouillage des dépendances

`uv.lock` est le **fichier de verrouillage** qui garantit la reproductibilité :

```
requirements.txt  vs  uv.lock

requirements.txt :
  ✅ Simple à lire
  ❌ Doit être géré manuellement
  ❌ Ne gère pas les dépendances transitives
  ❌ Peut avoir des conflits non détectés

uv.lock :
  ✅ Géré automatiquement par uv
  ✅ Contient TOUTES les dépendances (directes + transitives)
  ✅ Hash de chaque package pour la sécurité
  ✅ Compatible multi-plateformes
  ✅ Résolution de conflits garantie
```

**⚠️ Ne jamais modifier `uv.lock` manuellement !** Utilise toujours `uv add` / `uv remove`.

## 8️⃣ Cas Pratique : Projet Data Science avec uv

Voici comment structurer un **projet data science** de bout en bout avec `uv` :

### 📁 Structure recommandée

```
projet_analyse_ventes/
├── .python-version        ← Python 3.11 (fixé par uv)
├── pyproject.toml         ← Configuration et dépendances
├── uv.lock                ← Versions exactes figées
├── .gitignore             ← Exclure .venv/
├── README.md
├── .venv/                 ← Environnement virtuel (ignoré par Git)
├── data/
│   ├── raw/
│   └── processed/
├── notebooks/
│   ├── 01_exploration.ipynb
│   └── 02_modélisation.ipynb
└── src/
    └── pipeline.py
```

### 🏁 Du début à la fin

```bash
# 1️⃣ Créer le projet
uv init projet_analyse_ventes
cd projet_analyse_ventes

# 2️⃣ Définir la version Python
uv python pin 3.11

# 3️⃣ Ajouter les dépendances data
uv add pandas numpy scikit-learn plotly jupyter

# 4️⃣ Ajouter les outils de dev
uv add --dev pytest black ruff

# 5️⃣ Lancer Jupyter (uv s'occupe de l'activation !)
uv run jupyter notebook

# 6️⃣ Partager le projet
git add pyproject.toml uv.lock .python-version
git commit -m "chore: initialisation projet avec uv"

# ──────── Un collègue clone le projet ────────

# 7️⃣ Reproduire EXACTEMENT le même environnement
git clone <url_du_projet>
cd projet_analyse_ventes
uv sync                     # ← Installe EXACTEMENT les mêmes versions
uv run jupyter notebook     # ← Prêt à travailler !
```

### 💡 Pourquoi `uv sync` garantit la reproductibilité ?

`uv sync` lit `uv.lock` et installe **exactement** les mêmes versions, avec les mêmes hashs cryptographiques — garantissant que tu as le **même binaire** que ton collègue.

## 9️⃣ Bonnes Pratiques Récapitulatives

### ✅ Ce qu'il faut TOUJOURS faire

```
1. 🔒 Toujours utiliser un environnement virtuel par projet
   → python -m venv venv  ou  uv init

2. 📌 Figer les versions des packages
   → pip freeze > requirements.txt  ou  uv add (auto)

```

```
2. ❌ Utiliser des versions non-fixées en production

3. ❌ Partager le dossier venv/ par email ou git
   → Chemins absolus, dépendant du système

4. ❌ Modifier uv.lock manuellement
   → Utiliser uv add / uv remove

5. ❌ Ignorer pip check
   → Peut révéler des conflits silencieux
```

````python
# Résumé interactif : afficher l'environnement actuel
import sys
import os
import importlib.metadata

print("=" * 60)
print("  📊 RÉSUMÉ DE L'ENVIRONNEMENT PYTHON ACTUEL")
print("=" * 60)

# Infos Python
print(f"\n🐍 Python version    : {sys.version.split()[0]}")
print(f"📍 Interpréteur      : {sys.executable}")

# Environnement virtuel
est_venv = sys.prefix != sys.base_prefix
print(f"\n🔒 Environnement virtuel : {'✅ Actif' if est_venv else '⚠️  Non actif (Python global)'}")
if est_venv:
    print(f"   📁 Dossier : {sys.prefix}")

# Packages installés
packages_data = ["pandas", "numpy", "scikit-learn", "matplotlib", "plotly", "jupyter"]
print("\n📦 Packages data science :")
for pkg in packages_data:
    try:
        version = importlib.metadata.version(pkg)
        print(f"   ✅ {pkg:<20} v{version}")
    except importlib.metadata.PackageNotFoundError:
        print(f"   ❌ {pkg:<20} Non installé")

print("\n" + "=" * 60)
````
